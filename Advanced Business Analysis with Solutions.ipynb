# Basic Changes and Modifications to our Data Set

```python
# extracting different features from datetime column

df['Date'] = df['CREATED_DATE minus Hour'].dt.strftime('%d')
df['Month'] = df['CREATED_DATE minus Hour'].dt.strftime('%m')
df['Year'] = df['CREATED_DATE minus Hour'].dt.strftime('%Y')
df['Day'] = df['CREATED_DATE minus Hour'].dt.day_name()
df['Hour'] = df['CREATED_DATE'].dt.hour
```

```python
df.head()
```

```python
exchange_rates = {
    'EUR': 1.0,
    'GBP': 1.17,
}

# Convert prices to EUR based on the currency column
df['Amount In EUR'] = df.apply(lambda row: row['AMOUNT'] * exchange_rates[row['CURRENCY']], axis=1)
```

From here on new monthly user means the users who have no transactions prior. So if a user has a transaction in Jan 2016.
His transaction in feb 2016 is considered as a repeat customer and not a new one. 
Through this Section we are Trying to find the MONTH ON MONTH new user acquisition.
New user acquisition is a very important parameter for any business and describes a lot about business growth
